# Cleaning the data

```python
print(('_')*120)
print('\t\t\t\t*********************ANALYSIS*********************\n\t\t\t\tWe can clearly see that there are 2 columns with null values.\n\t\t\t\tAnd 2 columns with incorrect Data type.\n\t\t\t\tWe will be correcting the datatypes first as they may have hidden Null values.')
print(('_')*120)
val.info()
```

```python
val['KAST']=val['KAST'].str.replace('%','')
```

```python
val['Prize']=val['Prize'].str.replace('$','')
val['Prize']=val['Prize'].str.replace(',','')
```

```python
val['Death'].replace('\tnot defined',np.nan,inplace=True)
```

```python
val['Kill'].replace('\tnot defined',np.nan,inplace=True)
val['Kill'].replace('not defined',np.nan,inplace=True)
```

```python
val['KAST']=val['KAST'].astype(int)
```

```python
val['Prize']=val['Prize'].astype(float)
```

```python
val['Death']=val['Death'].astype(float)
```

```python
val['Kill']=val['Kill'].astype(float)
```

```python
print(('_')*120)
print('\t\t\t\t*********************ANALYSIS*********************\n\t\t\t\tAs you can see now all the columns Dtypes are Corrected.\n\t\t\t\tAnd hidden null values have been found.\n\t\t\t\tNow we can proceed to solving the Null values')
print(('_')*120)
val.info()
```

# Finding Total no. of Null values in columns

```python
print(('_')*120)
print('\t\t\t\t*********************ANALYSIS*********************\n\t\t\t\tThere are total of 4 null values in K/D and HS% columns.\n\t\t\t\t and 7 null valuses in Kill and 6 in Death \n\t\t\t\tSo we need to get rid of those null values  ')
print(('_')*120)
val.isnull().sum()
```

# Finding % of Null values

```python
print(('_')*120)
print('\t\t\t\t*********************ANALYSIS*********************\n\t\t\t\t As the null values are 10% of the columns K/D and HS %.\n\t\t\t\tand 17% and 15% of Kill and Death \n\t\t\t\tSo we need to fill them with average of respected columns')
print(('_')*120)
val.isnull().sum()/len(val)*100
```

```python
x=(val['K/D'].mean())
val['K/D'].fillna(x,inplace=True)
```

```python
H=(val['HS %'].mean())
val['HS %'].fillna(H,inplace=True)
```

```python
K=(val['Kill'].mean())
val['Kill'].fillna(K,inplace=True)
val['Death']=val['Death'].astype(float)
```

```python
D=(val['Death'].mean())
val['Death'].fillna(D,inplace=True)
val['Death']=val['Death'].astype(int)
```

```python
print(('_')*120)
print('\t\t\t\t*********************ANALYSIS*********************\n\t\t\t\t As you can see all Null values have been handled.\n\t\t\t\tThe Dataset is finally Cleaned.\n\t\t\t\t We can now proceed with Analysis. ')
print(('_')*120)
val.info()
```

### Q2 Which teams had the highest win & lose rates?

```python

val['Win Rate'] = val['Rounds Win'] / val['Rounds Played']
team_win_rates = val.groupby('Team')['Win Rate'].mean()
Teams = team_win_rates.sort_values(ascending=False)
print('Teams win rate')
Teams

```

```python
val['Lose Rate'] = val['Rounds Lose'] / val['Rounds Played']
team_lose_rates = val.groupby('Team')['Lose Rate'].mean()
Teams = team_lose_rates.sort_values(ascending=False)
print('teams lose rate')
Teams
```

```python
print(("_")*120)
print("\t\t\t\t**************ANALYSIS***************\n\n\t\t\t The Team with the Higihest Win Rate of the Competition is Loud with a Win Rate of 0.698%.\n\t\t\t The team with the Lowest Win Rate is Xset with a Win Rate of 0.388%.")
print(("_")*120)
y=val['Win Rate']
y2=val['Lose Rate']
x=val['Team']
plt.plot(x,y,marker='o',color='green',label='Win')
plt.plot(x,y2,marker='o',color='red',label='Lose')
plt.xlabel('Name of Teams')
plt.ylabel('Win & Lose Rates')
plt.legend()
plt.grid()
plt.xticks(rotation=90)
plt.show()
```
