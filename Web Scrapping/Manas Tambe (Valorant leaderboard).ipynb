```python
#importing required Libraries
import requests                  #to send the request to the URL
from bs4 import BeautifulSoup    #to get the content in the form of HTML
import pandas as pd              #to create dataframe
```

```python
#assigning the URL with variable name url
url='https://tracker.gg/valorant/leaderboards/ranked/all/default?page=1'
```

```python
#request allow you to send HTTP request
page=requests.get(url)
```

```python
page
```

```python
Soup=BeautifulSoup(page.text)
Soup.prettify
```

```python
#storing the meaningfull required datas in the variables
table=Soup.find('table')
```

```python
table=table.find('tbody')
table
```

```python
rows=table.find_all('tr')
rows
```

```python
#Using for loop to append data
columns=[]
for i in rows:
    cols=i.find_all('td')
    print(cols)
    players_info=[c.text for c in cols]
    print(players_info)
    columns.append (players_info)
print(columns)
```

```python
#creating an empty list, so that we can append the values and calling one by one using for loop
Rank=[]
Player=[]
Ranked_Rating=[]
Tier=[]
Wins=[]
for i in columns:
    Rank.append(i[0])
    Player.append(i[1])
    Ranked_Rating.append(i[2])
    Tier.append(i[3])
    Wins.append(i[4])
```

```python
#Storing all inputs in a variable
Data={'Ranks':Rank,'Player':Player,'Ranked_Rating':Ranked_Rating,'Tier':Tier,'Wins':Wins}
```

```python
#Saving the final data in the attribute:
Leaderbord=pd.DataFrame(Data)
```

```python
#Creating a csv file
Leaderbord.to_csv('Valorant.csv')
```
